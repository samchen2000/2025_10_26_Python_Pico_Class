# Python 基礎語法與名詞說明

## 1. 變數（Variable）
**變數**就像一個盒子，用來存放資料。你可以給這個盒子取一個名字，然後把資料放進去。


```python
# 變數的定義和使用
# 語法：變數名稱 = 值

# 範例 1：數字變數
age = 20          # age 是變數名稱，20 是儲存的值
price = 99.5      # 可以是小數
quantity = 10     # 可以是整數

# 範例 2：字串變數（文字）
name = "張三"     # 用雙引號或單引號包起來的都是字串
message = '你好'

# 範例 3：使用變數
print(f"姓名：{name}，年齡：{age}")
print(f"價格：{price}，數量：{quantity}")
print(f"總價：{price * quantity}")
```

## 2. 資料型別（Data Type）
Python 中常見的資料型別有以下幾種：


```python
# 2.1 整數（int）- 沒有小數點的數字
num1 = 10
num2 = -5
num3 = 0

# 2.2 浮點數（float）- 有小數點的數字
price1 = 99.99
price2 = 3.14
price3 = -2.5

# 2.3 字串（str）- 文字資料
text1 = "Hello"
text2 = 'World'
text3 = "123"  # 注意：這是字串，不是數字！

# 2.4 布林值（bool）- 只有 True 或 False
is_student = True
is_working = False

# 2.5 列表（list）- 用方括號 [] 包起來，可以存放多個資料
fruits = ["蘋果", "香蕉", "橘子"]
numbers = [1, 2, 3, 4, 5]

# 使用 type() 函數查看變數的資料型別
print(f"num1 的型別：{type(num1)}")
print(f"price1 的型別：{type(price1)}")
print(f"text1 的型別：{type(text1)}")
print(f"is_student 的型別：{type(is_student)}")
print(f"fruits 的型別：{type(fruits)}")
```

## 3. 運算符（Operators）
運算符用來進行數學運算或比較


```python
# 3.1 算術運算符
a = 10
b = 3

print(f"加法：{a} + {b} = {a + b}")      # 加
print(f"減法：{a} - {b} = {a - b}")      # 減
print(f"乘法：{a} * {b} = {a * b}")      # 乘
print(f"除法：{a} / {b} = {a / b}")      # 除（結果是浮點數）
print(f"整除：{a} // {b} = {a // b}")    # 整除（只取整數部分）
print(f"餘數：{a} % {b} = {a % b}")      # 取餘數
print(f"次方：{a} ** {b} = {a ** b}")    # 次方運算（10 的 3 次方）

# 3.2 比較運算符（結果是 True 或 False）
x = 5
y = 10

print(f"{x} == {y}：{x == y}")  # 等於
print(f"{x} != {y}：{x != y}")  # 不等於
print(f"{x} > {y}：{x > y}")    # 大於
print(f"{x} < {y}：{x < y}")    # 小於
print(f"{x} >= {y}：{x >= y}")  # 大於等於
print(f"{x} <= {y}：{x <= y}")  # 小於等於
```

## 4. 字串（String）的操作
字串是 Python 中處理文字的重要資料型別


```python
# 4.1 字串連接
first_name = "張"
last_name = "三"
full_name = first_name + last_name  # 用 + 號連接字串
print(f"全名：{full_name}")

# 4.2 字串格式化（f-string）- 推薦的方式
name = "李四"
age = 25
message = f"我的名字是 {name}，今年 {age} 歲"
print(message)

# 4.3 字串的其他方法
text = "Hello Python"
print(f"轉大寫：{text.upper()}")        # 轉成大寫
print(f"轉小寫：{text.lower()}")        # 轉成小寫
print(f"字串長度：{len(text)}")         # 取得字串長度
print(f"第一個字：{text[0]}")           # 取得第一個字（索引從 0 開始）
print(f"前5個字：{text[0:5]}")          # 切片（取得前5個字）
```

## 6. 條件語句（Conditional Statements）
用來根據條件決定執行哪些程式碼


```python
# 6.1 if 語句（如果）
score = 56

if score >= 60:
    print("及格了！")  # 注意：縮排很重要，用 4 個空格或 Tab

# 6.2 if-else 語句（如果-否則）
score = 45

if score >= 60:
    print("及格了！")
else:
    print("不及格！")

# 6.3 if-elif-else 語句（如果-否則如果-否則）
score = 85

if score >= 90:
    print("優等")
elif score >= 80:  # elif 是 "else if" 的縮寫
    print("良好")
elif score >= 60:
    print("及格")
else:
    print("不及格")
```

## 7. 迴圈（Loops）
用來重複執行某些程式碼


```python
# 7.1 for 迴圈 - 遍歷列表
fruits = ["蘋果", "香蕉", "橘子"]

for fruit in fruits:
    print(f"我喜歡 {fruit}")

# 7.2 for 迴圈 - 使用 range() 函數
# range(5) 會產生 0, 1, 2, 3, 4
for i in range(5):
    print(f"數字：{i}")

# range(1, 6) 會產生 1, 2, 3, 4, 5（從1開始到5結束）
for i in range(1, 6):
    print(f"計數：{i}")

# 7.3 while 迴圈 - 當條件為 True 時繼續執行
count = 0
while count < 5:
    print(f"計數：{count}")
    count = count + 1  # 每次迴圈都把 count 加 1
```

## 8. 列表（List）的操作
列表用來存放多個資料


```python
# 8.1 建立列表
fruits = ["蘋果", "香蕉", "橘子"]
numbers = [1, 2, 3, 4, 5]
mixed = ["文字", 123, 3.14, True]  # 列表可以存放不同型別的資料

# 8.2 存取列表元素（索引從 0 開始）
print(f"第一個水果：{fruits[0]}")   # 索引 0 是第一個
print(f"第二個水果：{fruits[1]}")   # 索引 1 是第二個
print(f"最後一個水果：{fruits[-1]}") # 索引 -1 是最後一個

# 8.3 修改列表元素
fruits[0] = "西瓜"  # 把第一個元素改成 "西瓜"
print(f"修改後：{fruits}")

# 8.4 新增元素到列表
fruits.append("葡萄")  # 在最後面新增
print(f"新增後：{fruits}")

# 8.5 列表的長度
print(f"列表長度：{len(fruits)}")

# 8.6 遍歷列表
for fruit in fruits:
    print(fruit)
```

## 9. 函數（Function）
函數是一段可以重複使用的程式碼


```python
# 9.1 定義函數
# 語法：def 函數名稱(參數1, 參數2, ...):
def greet(name):
    """這是一個打招呼的函數"""
    return f"你好，{name}！"

# 9.2 呼叫函數
message = greet("張三")
print(message)

# 9.3 帶多個參數的函數
def add(a, b):
    """計算兩個數字的和"""
    return a + b

result = add(10, 20)
print(f"10 + 20 = {result}")

# 9.4 沒有回傳值的函數
def print_info(name, age):
    """印出個人資訊"""
    print(f"姓名：{name}，年齡：{age}")

print_info("李四", 25)
```

## 10. 綜合練習
讓我們用學過的語法來寫一個簡單的程式


```python
# 綜合範例：計算購物車總價

# 定義商品和價格
products = ["蘋果", "香蕉", "橘子"]
prices = [50, 30, 40]
quantities = [2, 3, 1]

# 計算總價
total = 0
for i in range(len(products)):
    product_total = prices[i] * quantities[i]
    total = total + product_total
    print(f"{products[i]}：{prices[i]} 元 x {quantities[i]} = {product_total} 元")

print(f"總共：{total} 元")

# 判斷是否有折扣
if total >= 200:
    discount = total * 0.1  # 10% 折扣
    final_total = total - discount
    print(f"折扣：{discount} 元")
    print(f"應付：{final_total} 元")
else:
    print(f"未達折扣門檻，應付：{total} 元")
```
