```python
chinese = int(input("國文分數"))
print(type(chinese))
english = int(input("英文分數"))
print(type(english))
math = int(input("數學分數"))
print(type(math))

total = chinese + english + math
print("總分 : ", total, "分", sep = "__")
average = round(total / 3, ndigits=2)
print("平均 : ", average, "分", sep="---")
```

```python
import random
score = random.randint(1,100)
presure = random.randint(30, 180)
dis = ["優", "甲", "乙", "丙", "丁"]
bol = ["很高", "偏高", "正常", "偏低", "很低"]
if score > 90:
    print(f"分數為  {score}, {dis[0]}")
elif score >= 80:
    print(f"分數為  {score}, {dis[1]}")
elif score >= 70:
    print(f"分數為  {score}, {dis[2]}")
elif score >= 60:
    print(f"分數為  {score}, {dis[3]}")
else :
    print(f"分數為  {score}, {dis[4]}")

if  score > 90 :
    print(f"分數為  {score}, {dis[0]}")
elif 90 >= score > 80:
    print(f"分數為  {score}, {dis[1]}")
elif 80 >= score > 70:
    print(f"分數為  {score}, {dis[2]}")
elif 70 >= score >= 60:
    print(f"分數為  {score}, {dis[3]}")
else :
    print(f"分數為  {score}, {dis[4]}")

if presure > 90 and presure < 120:
    print(f"血壓為 : {presure}, {bol[2]}")
elif presure <= 90 and presure > 55:
    print(f"分數為  {presure}, {bol[3]}")
elif presure <= 55:
    print(f"分數為  {presure}, {bol[4]}")
elif presure >= 120 and presure <= 150:
    print(f"分數為  {presure}, {bol[1]}")
else :
    print(f"分數為  {presure}, {bol[0]}")

if presure <= 90 or presure >= 120:
    print(f"血壓為 : {presure}, 不正常")
else :
    print(f"分數為  {presure}, 正常")    

if not(presure <= 90 or presure >= 120):
    print(f"血壓為 : {presure}, 正常")
else :
    print(f"分數為  {presure}, 不正常")      
```
